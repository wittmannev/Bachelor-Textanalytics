# Clustering der Entscheidungsgründe

## Datenvorbereitung

### Datenimport

Daten mit get_clustering_data aus DB laden.

```python
import sqlite3
import pandas as pd
import os.path

dbPath = "database/bachelor_umfrage.db"

def get_clustering_data(table_name="clustering"):
    df = pd.DataFrame()
    if os.path.isfile(dbPath):
        conn = sqlite3.connect(dbPath)
        sql = "SELECT * FROM " + table_name
        sql += " WHERE Gründe IS NOT NULL"
        #sql += " WHERE Gründe IS NOT NULL AND MINTFach == 'ja'"
        #sql += " WHERE Gründe IS NOT NULL AND MINTFach == 'nein'"
        df = pd.read_sql_query(sql, conn)
    else:
        print("Datenbank oder Tabelle konnte nicht gefunden werden.")
    return df

reason_df = get_clustering_data(table_name="clustering")
reason_df.sample(5)
```

### Datenaufbereitung

```python
import spacy

nlp = spacy.load('de')

# define space nlp with just the tagger (includes lemmatizer)
nlp_tagger = spacy.load('de', disable=['parser', 'ner'])
```

Beschränkung auf Nomen für das Clustering.

```python
import string

def get_stopwords():
    with open("stopwords_de.txt", encoding="utf-8") as file:
        stopwords = file.read().splitlines()
        stopwords += ['studium', 'studiengang', 'mint']
    return set(stopwords)

stopwords = get_stopwords()
stopwords = stopwords.union(set(string.punctuation))

def pos_tokenizer(text):
    spacy_doc = nlp_tagger(text)
    pos_list=['NOUN', 'PROPN']
    
    tokens = [token.lemma_ for token in spacy_doc if token.pos_ in pos_list]
    tokens = [token for token in tokens if token.lower() not in stopwords]
    return tokens
```

Tokenisierung mit anschließender Speicherung in das Dataframe.

```python
reason_df['tokens'] = reason_df['Gründe'].map(lambda t: ' '.join(pos_tokenizer(t)))
```

```python
pd.set_option('max_colwidth', -1)
reason_df[['Gründe', 'tokens']].sample(5)
pd.reset_option('max_colwidth')
```

### Vektorisierung

Ermittlung der TF- und TF-IDF-Matrix

```python
from sklearn.feature_extraction.text import CountVectorizer

# TF-Vectorizer
count_vect = CountVectorizer(lowercase=False, min_df=10, max_df=0.5)
X_train_tf = count_vect.fit_transform(reason_df.tokens)
print("%d x %d Count-Matrix (= BoW = TF)" % X_train_tf.shape)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

# TF-IDF Vectorizer
tfidf_transformer = TfidfTransformer(sublinear_tf=True, norm="l2")
X_train_tfidf = tfidf_transformer.fit_transform(X_train_tf)

print("%d x %d TF-IDF-Matrix" % X_train_tfidf.shape)
```

Wert wird zunächst auf TD-IDF-Matrix gesetzt. Um später andere Matrix zu testen wird dieser Wert geändert.

```python
X_train = X_train_tfidf
```

### Häufigste Wörter

Ausgabe der 10 häufigsten Wörter je Cluster.

```python
feature_names = count_vect.get_feature_names()
counts_df = pd.DataFrame(X_train.toarray(), columns=feature_names)
```

```python
def get_most_important_words(counts_df, labels):
    # Kopie des DataFrames anlegen, Werte werden manipuliert
    d = counts_df.copy()
    
    d["cluster"] = labels
    # Gruppierung nach Clusterlabel und Summierung der Häufigkeiten
    d = d.groupby(["cluster"]).sum()
    # Transponierung, jetzt sind die Cluster die Spalten und die Terme in den Zeilen
    dt = d.T
    
    # dt = dt.applymap(lambda x: x if x < 10000000 else 0)
    
    # Spalten-Namen vergeben: Cluster 0 ... k-1
    dt.columns = ["Cluster %d" % label for label in np.unique(labels)]
    
    # Ausgabe der 10 häufigsten Worte je Cluster
    for col in dt.columns: 
        # sort_values sortiert nach einer Spalte
        tmp = dt.sort_values(by=col, ascending=False)
        tmp = tmp[[col]]
        display(tmp.head(10))
```

```python
get_most_important_words(counts_df, k_means.labels_)
```

## Evakluation der Ergebnisse

Um zu bewerten, wie gut das Clustering ist werden im folgenden verschiedene Gütemaße berechnet.

### Intra-Cluster-Variance

```python
# Erstelle Matrix, die für jedes Dokument den Vektor 
# des entsprechenden Centroids enthält
X_centroids = k_means.cluster_centers_[k_means.labels_]
# ICV errechnet sich dann aus (X-X_centroids)**2
```

```python
def icv_score(X, X_centroids):
    D_squared = np.square(X-X_centroids)
    return np.sum(np.sum(D_squared, axis=1))
```

```python
icv_score(X_train, X_centroids)
```

ICV groß => Cluster sind nicht kompakt

### Silhouette-Score

```python
from sklearn.metrics import silhouette_score

silhouette_score(X_train.toarray(), k_means.labels_, metric="euclidean")
```

Silhouette-Score nahe 0 => schlecht seperierbare Cluster

### Silhouette Plot

```python
def visualize_silhouette(X, cluster_labels, cluster_centers, n_clusters):
    import numpy as np
    from sklearn.metrics import silhouette_samples, silhouette_score
    import matplotlib.cm as cm
    import matplotlib.pyplot as plt
    from sklearn.decomposition import PCA

    pca = PCA(n_components=2)
    pc = pca.fit_transform(X)
    centers_transformed = pca.transform(cluster_centers)

    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # Der Silhouette Plot ist der erste Subplot
    # Der Silhouette Score kann zwar Werte zwischen -1 und 1 annehmen,
    # nachdem der Score in diesem Beispiel jedoch sehr niedrig ist, wird
    # der Wertebereich der Grafik zur besseren Darstellbarkeit auf [-0.05, 0.15] eingeschränkt
    ax1.set_xlim([-0.05, 0.15])
    
    # Fügt Leerraum zwischen den Silhouette Plots für die einzelnen Cluster ein
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Berechnung des Silhouette Scores
    silhouette_score = silhouette_score(X, cluster_labels)
    print("Silhouette Score für n_clusters = {}: {}".format(n_clusters, silhouette_score))

    # Berechnet den Silhouette Score für jedes der 1.000 Samples einzeln 
    # (liefert Array mit 1.000 Scores)   
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregation der einzelnen Scores für cluster i und absteigende Sortierung
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.tab20c(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Setzt das Clusterlabel mittig links and den jeweiligen Silhouette Plot
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        
        # Berechnet einen neuen Wert für y_lower für den nächsten Plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("Silhouette Plot je Cluster.")
    ax1.set_xlabel("Silhouette Koeffizient")
    ax1.set_ylabel("Cluster Label")

    # Vertikale Linie, die den Mittelwert über alle Werte (den eigentlichen Silhouette Score) angibt
    ax1.axvline(x=silhouette_score, color="red", linestyle="--")

    ax1.set_yticks([])  # Y-Achsenbezeichnungen entfernen
    ax1.set_xticks([-0.05, 0, 0.05, 0.1, 0.15])


    # Zweiter Plot (rechts), der die Cluster mit Hilfe von PCA zweidimensional darstellt
    colors = cm.tab20c(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(pc[:, 0], pc[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling der Cluster im rechten Plot
    ax2.scatter(centers_transformed[:, 0], centers_transformed[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers_transformed):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("Geclusterte Daten mit zwei Hauptkomponenten (PCA)")
    ax2.set_xlabel("Erste Hauptkomponente")
    ax2.set_ylabel("Zweite Hauptkomponente")

    plt.suptitle(("Silhouette Analyse mit n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold');
```

```python
X = X_train.toarray()
cluster_labels = k_means.labels_
cluster_centers = k_means.cluster_centers_
n_clusters = k_means.n_clusters
```

```python
visualize_silhouette(X, cluster_labels, cluster_centers, n_clusters)
```

Nur wenige Datensätze mit negativem Score -> nur wenige falsch zugeordnet. Ein Cluster viel größer als die anderen -> vermutlich nicht die Optimale Clusteranzahl.

### Optimierung der Clusteranzahl

```python
def find_opt_k(X, range_n_clusters):
    import pandas as pd 
    import numpy as np
    import matplotlib.pyplot as plt
    from sklearn import metrics
    from sklearn.cluster import KMeans
    from sklearn.decomposition import PCA
    
    plt.rcParams["figure.figsize"] = (16, 9)
    plt.style.use("seaborn-whitegrid")
    
    plot_count = 1
    n_components = 2
    
    # Durchführung PCA
    pca = PCA(n_components=n_components)
    principal_components = pca.fit_transform(X.toarray())
    cols = []
    for i in range(n_components):
        cols.append("component" + str(i))
    pca_df = pd.DataFrame(data = principal_components, columns = cols)
    
    for n_clusters in range_n_clusters:
        
        # Durchführung K-Means für die aktuelle Clusteranzahl
        k_means = KMeans(n_clusters=n_clusters)
        k_means.fit(X)
        labels = k_means.labels_
        
        # Plot der Ergebnisse in einem 3x3 Raster
        plt.subplot(3,3,plot_count)
        plt.scatter(pca_df.iloc[:,0], pca_df.iloc[:,1], c=labels, cmap="tab20c", s=7)
        plt.title("n_clusters = " + str(n_clusters))
        
        # Berechnung der Performancemaße
        silhouette_score = metrics.silhouette_score(X.toarray(), labels)
        harabaz = metrics.calinski_harabaz_score(X.toarray(), labels)

        print("{:d} Cluster: Silhouette = {:.6f}, Calinski-Harabaz = {:.2f}".format(n_clusters, silhouette_score, harabaz))
        plot_count += 1
```

```python
%matplotlib inline
%config InlineBackend.figure_format = "retina"

# Vorgegebene Clusteranzahlen
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]

find_opt_k(X_train_tfidf, range_n_clusters)
```

## Datenvorbereitung

### Datenimport

Daten mit get_clustering_data aus DB laden.

```python
import sqlite3
import pandas as pd
import os.path

dbPath = "database/bachelor_umfrage.db"

def get_clustering_data(table_name="clustering"):
    df = pd.DataFrame()
    if os.path.isfile(dbPath):
        conn = sqlite3.connect(dbPath)
        sql = "SELECT * FROM " + table_name
        sql += " WHERE Verbesserungsvorschläge IS NOT NULL"
        df = pd.read_sql_query(sql, conn)
    else:
        print("Datenbank oder Tabelle konnte nicht gefunden werden.")
    return df

suggestions_df = get_clustering_data(table_name="clustering")
suggestions_df.sample(5)
```

### Datenaufbereitung

```python
import spacy

nlp = spacy.load('de')

# define space nlp with just the tagger (includes lemmatizer)
nlp_tagger = spacy.load('de', disable=['parser', 'ner'])
```

Beschränkung auf Nomen für das Clustering.

```python
import string

def get_stopwords():
    with open("stopwords_de.txt", encoding="utf-8") as file:
        stopwords = file.read().splitlines()
        stopwords += ['studium', 'studiengang', 'mint-fächer']
    return set(stopwords)

stopwords = get_stopwords()
stopwords = stopwords.union(set(string.punctuation))

def pos_tokenizer(text):
    spacy_doc = nlp_tagger(text)
    pos_list=['NOUN', 'PROPN']
    
    tokens = [token.lemma_ for token in spacy_doc if token.pos_ in pos_list]
    tokens = [token for token in tokens if token.lower() not in stopwords]
    return tokens
```

Tokenisierung mit anschließender Speicherung in das Dataframe.

```python
suggestions_df['tokens'] = suggestions_df['Verbesserungsvorschläge'].map(lambda t: ' '.join(pos_tokenizer(t)))
```

```python
pd.set_option('max_colwidth', -1)
suggestions_df[['Verbesserungsvorschläge', 'tokens']].sample(5)
pd.reset_option('max_colwidth')
```

### Vektorisierung

Ermittlung der TF- und TF-IDF-Matrix

```python
from sklearn.feature_extraction.text import CountVectorizer

# TF-Vectorizer
count_vect = CountVectorizer(lowercase=False, min_df=10, max_df=0.5)
X_train_tf = count_vect.fit_transform(suggestions_df.tokens)
print("%d x %d Count-Matrix (= BoW = TF)" % X_train_tf.shape)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

# TF-IDF Vectorizer
tfidf_transformer = TfidfTransformer(sublinear_tf=True, norm="l2")
X_train_tfidf = tfidf_transformer.fit_transform(X_train_tf)

print("%d x %d TF-IDF-Matrix" % X_train_tfidf.shape)
```

Wert wird zunächst auf TD-IDF-Matrix gesetzt. Um später andere Matrix zu testen wird dieser Wert geändert.

```python
X_train = X_train_tfidf
```

Ausgabe der 10 häufigsten Wörter je Cluster.

```python
feature_names = count_vect.get_feature_names()
counts_df = pd.DataFrame(X_train.toarray(), columns=feature_names)
```

```python
def get_most_important_words(counts_df, labels):
    # Kopie des DataFrames anlegen, Werte werden manipuliert
    d = counts_df.copy()
    
    d["cluster"] = labels
    # Gruppierung nach Clusterlabel und Summierung der Häufigkeiten
    d = d.groupby(["cluster"]).sum()
    # Transponierung, jetzt sind die Cluster die Spalten und die Terme in den Zeilen
    dt = d.T
    
    # dt = dt.applymap(lambda x: x if x < 10000000 else 0)
    
    # Spalten-Namen vergeben: Cluster 0 ... k-1
    dt.columns = ["Cluster %d" % label for label in np.unique(labels)]
    
    # Ausgabe der 10 häufigsten Worte je Cluster
    for col in dt.columns: 
        # sort_values sortiert nach einer Spalte
        tmp = dt.sort_values(by=col, ascending=False)
        tmp = tmp[[col]]
        display(tmp.head(10))
```

## Evakluation der Ergebnisse

Um zu bewerten, wie gut das Clustering ist werden im folgenden verschiedene Gütemaße berechnet.

### Intra-Cluster-Variance

```python
# Erstelle Matrix, die für jedes Dokument den Vektor 
# des entsprechenden Centroids enthält
X_centroids = k_means.cluster_centers_[k_means.labels_]
# ICV errechnet sich dann aus (X-X_centroids)**2
```

```python
def icv_score(X, X_centroids):
    D_squared = np.square(X-X_centroids)
    return np.sum(np.sum(D_squared, axis=1))
```

```python
icv_score(X_train, X_centroids)
```

ICV groß => Cluster sind nicht kompakt aber besser als bei vorherigem Clustering.

### Silhouette-Score

```python
from sklearn.metrics import silhouette_score

silhouette_score(X_train.toarray(), k_means.labels_, metric="euclidean")
```

Silhouette-Score nahe 0 => schlecht seperierbare Cluster. Aber besser als bei vorherigem Clustering.

### Silhouette Plot

```python
def visualize_silhouette(X, cluster_labels, cluster_centers, n_clusters):
    import numpy as np
    from sklearn.metrics import silhouette_samples, silhouette_score
    import matplotlib.cm as cm
    import matplotlib.pyplot as plt
    from sklearn.decomposition import PCA

    pca = PCA(n_components=2)
    pc = pca.fit_transform(X)
    centers_transformed = pca.transform(cluster_centers)

    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # Der Silhouette Plot ist der erste Subplot
    # Der Silhouette Score kann zwar Werte zwischen -1 und 1 annehmen,
    # nachdem der Score in diesem Beispiel jedoch sehr niedrig ist, wird
    # der Wertebereich der Grafik zur besseren Darstellbarkeit auf [-0.05, 0.15] eingeschränkt
    ax1.set_xlim([-0.05, 0.15])
    
    # Fügt Leerraum zwischen den Silhouette Plots für die einzelnen Cluster ein
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Berechnung des Silhouette Scores
    silhouette_score = silhouette_score(X, cluster_labels)
    print("Silhouette Score für n_clusters = {}: {}".format(n_clusters, silhouette_score))

    # Berechnet den Silhouette Score für jedes der 1.000 Samples einzeln 
    # (liefert Array mit 1.000 Scores)   
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregation der einzelnen Scores für cluster i und absteigende Sortierung
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.tab20c(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Setzt das Clusterlabel mittig links and den jeweiligen Silhouette Plot
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        
        # Berechnet einen neuen Wert für y_lower für den nächsten Plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("Silhouette Plot je Cluster.")
    ax1.set_xlabel("Silhouette Koeffizient")
    ax1.set_ylabel("Cluster Label")

    # Vertikale Linie, die den Mittelwert über alle Werte (den eigentlichen Silhouette Score) angibt
    ax1.axvline(x=silhouette_score, color="red", linestyle="--")

    ax1.set_yticks([])  # Y-Achsenbezeichnungen entfernen
    ax1.set_xticks([-0.05, 0, 0.05, 0.1, 0.15])


    # Zweiter Plot (rechts), der die Cluster mit Hilfe von PCA zweidimensional darstellt
    colors = cm.tab20c(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(pc[:, 0], pc[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling der Cluster im rechten Plot
    ax2.scatter(centers_transformed[:, 0], centers_transformed[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers_transformed):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("Geclusterte Daten mit zwei Hauptkomponenten (PCA)")
    ax2.set_xlabel("Erste Hauptkomponente")
    ax2.set_ylabel("Zweite Hauptkomponente")

    plt.suptitle(("Silhouette Analyse mit n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold');
```

```python
X = X_train.toarray()
cluster_labels = k_means.labels_
cluster_centers = k_means.cluster_centers_
n_clusters = k_means.n_clusters
```

```python
visualize_silhouette(X, cluster_labels, cluster_centers, n_clusters)
```

### Optimierung der Clusteranzahl

```python
def find_opt_k(X, range_n_clusters):
    import pandas as pd 
    import numpy as np
    import matplotlib.pyplot as plt
    from sklearn import metrics
    from sklearn.cluster import KMeans
    from sklearn.decomposition import PCA
    
    plt.rcParams["figure.figsize"] = (16, 9)
    plt.style.use("seaborn-whitegrid")
    
    plot_count = 1
    n_components = 2
    
    # Durchführung PCA
    pca = PCA(n_components=n_components)
    principal_components = pca.fit_transform(X.toarray())
    cols = []
    for i in range(n_components):
        cols.append("component" + str(i))
    pca_df = pd.DataFrame(data = principal_components, columns = cols)
    
    for n_clusters in range_n_clusters:
        
        # Durchführung K-Means für die aktuelle Clusteranzahl
        k_means = KMeans(n_clusters=n_clusters)
        k_means.fit(X)
        labels = k_means.labels_
        
        # Plot der Ergebnisse in einem 3x3 Raster
        plt.subplot(3,3,plot_count)
        plt.scatter(pca_df.iloc[:,0], pca_df.iloc[:,1], c=labels, cmap="tab20c", s=7)
        plt.title("n_clusters = " + str(n_clusters))
        
        # Berechnung der Performancemaße
        silhouette_score = metrics.silhouette_score(X.toarray(), labels)
        harabaz = metrics.calinski_harabaz_score(X.toarray(), labels)

        print("{:d} Cluster: Silhouette = {:.6f}, Calinski-Harabaz = {:.2f}".format(n_clusters, silhouette_score, harabaz))
        plot_count += 1
```

```python
%matplotlib inline
%config InlineBackend.figure_format = "retina"

# Vorgegebene Clusteranzahlen
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]

find_opt_k(X_train_tfidf, range_n_clusters)
```
