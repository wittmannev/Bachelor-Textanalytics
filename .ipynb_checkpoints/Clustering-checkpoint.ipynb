# Datenvorbereitung

## Datenimport

Daten mit get_clustering_data aus DB laden.

```python
import sqlite3
import pandas as pd
import os.path

dbPath = "database/bachelor_umfrage.db"

def get_clustering_data(table_name="clustering"):
    df = pd.DataFrame()
    if os.path.isfile(dbPath):
        conn = sqlite3.connect(dbPath)
        sql = "SELECT * FROM " + table_name
        sql += " WHERE Gründe IS NOT NULL"
        df = pd.read_sql_query(sql, conn)
    else:
        print("Datenbank oder Tabelle konnte nicht gefunden werden.")
    return df

reason_df = get_clustering_data(table_name="clustering")
reason_df.sample(5)
```

## Datenaufbereitung

```python
import spacy

nlp = spacy.load('de')

# define space nlp with just the tagger (includes lemmatizer)
nlp_tagger = spacy.load('de', disable=['parser', 'ner'])
```

Beschränkung auf Nomen für das Clustering.

```python
import string

def get_stopwords():
    with open("stopwords_de.txt", encoding="utf-8") as file:
        stopwords = file.read().splitlines()
    return set(stopwords)

stopwords = get_stopwords()
stopwords = stopwords.union(set(string.punctuation))

def pos_tokenizer(text):
    spacy_doc = nlp_tagger(text)
    pos_list=['NOUN', 'PROPN']
    
    tokens = [token.lemma_ for token in spacy_doc if token.pos_ in pos_list]
    tokens = [token for token in tokens if token.lower() not in stopwords]
    return tokens
```

Tokenisierung mit anschließender Speicherung in das Dataframe.

```python
reason_df['tokens'] = reason_df['Gründe'].map(lambda t: ' '.join(pos_tokenizer(t)))
```

```python
pd.set_option('max_colwidth', -1)
reason_df[['Gründe', 'tokens']].head()
pd.reset_option('max_colwidth')
```

## Vektorisierung

Ermittlung der TF- und TF-IDF-Matrix

```python
from sklearn.feature_extraction.text import CountVectorizer

# TF-Vectorizer
count_vect = CountVectorizer(lowercase=False, min_df=10, max_df=0.5)
X_train_tf = count_vect.fit_transform(reason_df.tokens)
print("%d x %d Count-Matrix (= BoW = TF)" % X_train_tf.shape)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

# TF-IDF Vectorizer
tfidf_transformer = TfidfTransformer(sublinear_tf=True, norm="l2")
X_train_tfidf = tfidf_transformer.fit_transform(X_train_tf)

print("%d x %d TF-IDF-Matrix" % X_train_tfidf.shape)
```

Wert wird zunächst auf TD-IDF-Matrix gesetzt. Um später andere Matrix zu testen wird dieser Wert geändert.

```python
X_train = X_train_tfidf
```

# Clustering mit k-means

In diesem Kapitel sollen mit Hilfe des Clustering mit k-means die Gründe für oder gegen die Wahl von MINT-Fächern analysiert werden.

## Initialisierung & Training

Setzen der Anzahl der Cluster

```python
n_clusters = 5
```

```python
from sklearn.cluster import KMeans

k_means = KMeans(n_clusters=n_clusters, init='k-means++')
k_means.fit(X_train)
```

## Clusterzugehörigkeit der Trainingsdokumente

Aufteilung der Dokumente zu den einzelnen Clustern

```python
import numpy as np

cluster_count = np.unique(k_means.labels_, return_counts=True)
cluster_count
```

Anfügen der Clusterzuordnung an das Dataframe

```python
reason_df['cluster'] = k_means.labels_
```

```python
reason_df.groupby(['cluster']).agg({'AntwortID':'count'}).plot(kind='bar', figsize=(6,4));
```

## Häufigste Wörter

Ausgabe der 10 häufigsten Wörter je Cluster.

```python
feature_names = count_vect.get_feature_names()
counts_df = pd.DataFrame(X_train.toarray(), columns=feature_names)
```

```python
def get_most_important_words(counts_df, labels):
    # Kopie des DataFrames anlegen, Werte werden manipuliert
    d = counts_df.copy()
    
    d["cluster"] = labels
    # Gruppierung nach Clusterlabel und Summierung der Häufigkeiten
    d = d.groupby(["cluster"]).sum()
    # Transponierung, jetzt sind die Cluster die Spalten und die Terme in den Zeilen
    dt = d.T
    
    # dt = dt.applymap(lambda x: x if x < 10000000 else 0)
    
    # Spalten-Namen vergeben: Cluster 0 ... k-1
    dt.columns = ["Cluster %d" % label for label in np.unique(labels)]
    
    # Ausgabe der 10 häufigsten Worte je Cluster
    for col in dt.columns: 
        # sort_values sortiert nach einer Spalte
        tmp = dt.sort_values(by=col, ascending=False)
        tmp = tmp[[col]]
        display(tmp.head(10))
```

```python
get_most_important_words(counts_df, k_means.labels_)
```

# Visualisierung der Ergebnisse

In diesem Kapitel sollen die Ergebnisse durch PCA mit 2 und 3 Komponenten dargestellt werden

## PCA mit 2 Komponenten

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = "retina"
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train.toarray())
```

```python
centers_pca = pca.transform(k_means.cluster_centers_)

# Darstellung
plt.rcParams["figure.figsize"] = (16, 9)
plt.style.use("seaborn-whitegrid")
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=k_means.labels_, cmap="tab20c", s=7)

# Darstellung der Clusterzentren
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], marker="*", c="#ffcc5c", s=400);
```

## PCA mit 3 Komponenten

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
%config InlineBackend.figure_format = "retina"
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_train.toarray())
centers_pca = pca.transform(k_means.cluster_centers_)
```

```python
plt.rcParams["figure.figsize"] = (16, 9)
plt.style.use("seaborn-whitegrid")
fig = plt.figure()
ax = Axes3D(fig)

# Panes transparent darstellen
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))

ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=k_means.labels_, cmap="tab20c")
# Clusterzentren
ax.scatter(centers_pca[:, 0], centers_pca[:, 1], centers_pca[:, 2], marker="*", c="#ffcc5c", s=1200);

# Perspektive einstellen
ax.view_init(20, 30)
plt.draw();
```
